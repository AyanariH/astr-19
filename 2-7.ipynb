```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)

def exp_function(x):
    return np.exp(x)

y = exp_function(x)
```

```python
y = exp_function(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = e^x$', color='blue')

plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.title("Plot of $y = e^x$")
plt.legend()

plt.savefig("exp-graph.pdf",bbox_inches="tight",facecolor="white")

plt.show()
```
