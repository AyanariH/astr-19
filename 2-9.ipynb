```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 random numbers uniformly distributed between 0 and 1
data = np.random.uniform(0, 1, 1000)

# Create a histogram with 100 bins
plt.figure(figsize=(8, 6))
plt.hist(data, bins=100, color='skyblue', edgecolor='black')

# Label the axes
plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Histogram of 1000 Uniformly Distributed Random Numbers')

# Save the plot as a PDF
plt.savefig("histogram.pdf",bbox_inches="tight",facecolor="white")
plt.show()
```
