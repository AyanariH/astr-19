```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rayleigh(scale=1.0, size=1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=100, color='skyblue', edgecolor='black')

plt.xlabel('x value')
plt.ylabel('y value')
plt.title('Histogram of 1000 Random Rayleigh Distribution Numbers')

plt.savefig("rayleigh-histo.pdf",bbox_inches="tight",facecolor="white")
plt.show()
```
